```python
import pandas as pd
```

```python
df = pd.read_excel('customer_info.xlsx')
```

```python
df.head(5)
```

```python
df1 = pd.read_excel('Digital Ops DaTA.xlsx')
```

```python
df1.head(5)
```

```python
# Rename the column in df1 for consistency before merging (optional)
df1.rename(columns={'Apon_code': 'Customer Code'}, inplace=True)

# Merge df1 with df based on 'Customer Code'
# This operation left joins df1 with df, so all records in df1 will be kept
# and those in df will be merged where the 'Customer Code' matches
df1 = df1.merge(df[['Customer Code', 'Credit Limit']], on='Customer Code', how='left')
```

```python
df1.head(5)
```

```python
df1.info()
```

```python
# Drop rows in df1 where 'Credit Limit' is NaN
df1 = df1.dropna(subset=['Credit Limit'])
```

```python
df1.head(5)
```

```python
df1.info()
```

```python
df1['Remaining Credit'] = df1['Credit Limit'] - df1['numCreditAmount']
```

```python
df1.head(5)
```

```python
df1 = df1[df1['Remaining Credit'] >= 1000]
```

```python
df1.head(5)
```

```python
df1.info()
```

```python
df1.to_excel('remaing_credit_data.xlsx')  # Default is index=True
print("Data exported to remaing_credit_data successfully!")
```
